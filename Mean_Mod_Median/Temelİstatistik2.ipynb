```python
#Variance/Varyans
#Varyans,verilerin aritmetik ortalamadan sapmalarının karelerinin toplamının örnek sayısının 1 eksiğine bölümüdür.
#Aslında varyans : Standart sapmanın karekök alınmamış halidir.
```

```python
#Standard Deviation/ Standart Sapma
```

```python
import numpy as np
```

```python
#Random kullanarak sample bir veri setini kendimiz oluşturalım.
#Aylık gelir 5000TL Standart Deviation 2000TL olsun 1000 kişinin random aylık gelirini oluşturalım:

gelirler = np.random.normal(5000,2000,1000)
gelirler
```

```python
#Varyans Hesaplama
gelirler.var() 
```

```python
#Standart Sapma Hesaplama
gelirler.std()
```
